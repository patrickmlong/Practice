```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic1.csv')
```

```python
titanic.head()
```

```python
plt.hist(titanic.age.dropna(), bins = 40)
plt.show()
```

```python
sns.boxplot(titanic.age.dropna())
plt.show()
```

```python
sns.boxplot(titanic.age.dropna(), titanic.sex)
plt.show()
```

```python
sns.kdeplot(titanic.age.dropna(), shade = True)
plt.show()
```

```python
v= titanic.age.dropna()
plt.hist(v)
plt.show()
```

```python
v= titanic['age'].dropna()
plt.hist(v)
plt.show()
```

```python
sns.distplot(titanic.age.dropna(), 30)
plt.show()
```

```python
sns.violinplot(titanic.age.dropna(), titanic.sex)
plt.show()
```
